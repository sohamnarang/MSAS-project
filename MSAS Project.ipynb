```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
bundesliga1920 = pd.read_csv("bundesliga1920.csv")
clo1920 = pd.read_csv("champ.csv")
clo2021 = pd.read_csv("cl1920.csv")
premo2021 = pd.read_csv("prem2021.csv")
bundes1920 = bundesliga1920.head(4)
prem2021 = premo2021.head(4)
cl1920 = clo1920.head(4)
cl2021 = clo2021.head(4)

#xG-g, xga -a, xgd all per 90 call it off, def, xgd
```

```python
bundes1920["off_eff"] = (bundes1920["GF"] - bundes1920["xG"])/bundes1920["MP"]
bundes1920["def_eff"] = (bundes1920["xGA"] - bundes1920["GA"])/bundes1920["MP"]

cl1920["off_eff"] = (cl1920["GF"] - cl1920["xG"])/cl1920["MP"]
cl1920["def_eff"] = (cl1920["xGA"] - cl1920["GA"])/cl1920["MP"]

prem2021["off_eff"] = (prem2021["GF"] - prem2021["xG"])/prem2021["MP"]
prem2021["def_eff"] = (prem2021["xGA"] - prem2021["GA"])/prem2021["MP"]

cl2021["off_eff"] = (cl2021["GF"] - cl2021["xG"])/cl2021["MP"]
cl2021["def_eff"] = (cl2021["xGA"] - cl2021["GA"])/cl2021["MP"]
```

```python
x = np.arange(4)

y1 = bundes1920["off_eff"]
y2 = bundes1920["def_eff"]
y3 = bundes1920["xGD/90"]
width = 0.2
  
plt.bar(x-0.2, y1, width, color='tab:blue')
plt.bar(x, y2, width, color='tab:pink')
plt.bar(x+0.2, y3, width, color='tab:green')
plt.xticks(x, ['Bayern Munich', 'Dortmund', 'RB Leipzig', "M'Gladbach"])
plt.xlabel("Bundesliga 19/20 Teams")
plt.ylabel("Expected Metrics")
plt.legend(["off_eff", "def_off", "xGD/90"])
plt.axhline(y=0, color='k', linewidth = 0.5)
plt.show()
```
